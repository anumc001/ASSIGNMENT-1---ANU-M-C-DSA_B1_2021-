# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
plt.plot(data1["Classification"],data1["SepalLength cm"], color = 'r', marker = '*',ms = 10, mec = 'g')

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sapel Length of Diferent species", fontdict = font1)
plt.xlabel("Classification", fontdict = font2)
plt.ylabel("SepalLength cm", fontdict = font2)
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(x="Classification", y="SepalWidth cm", hue="Classification",data=data1)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sapel Width of Diferent species", fontdict = font1)
plt.xlabel("Classification", fontdict = font2)
plt.ylabel("SepalWidth cm", fontdict = font2)
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x="Classification", y="PetalLength cm",hue="Classification",data=data1)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("PetalLength of Diferent species", fontdict = font1)
plt.xlabel("Classification", fontdict = font2)
plt.ylabel("PetalLength cm", fontdict = font2)
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x="Classification", y="Petalwidth cm", hue="Classification",data=data1)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Petalwidth of Diferent species", fontdict = font1)
plt.xlabel("Classification", fontdict = font2)
plt.ylabel("Petalwidth cm", fontdict = font2)
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x="Classification", y="SepalLength cm",hue="Classification",data=data1)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("SepalLength of Diferent species", fontdict = font1)
plt.xlabel("Classification", fontdict = font2)
plt.ylabel("SepalLength cm", fontdict = font2)
```

```python
plt.figure(figsize=(10,10))
plt.hist(data1['Classification'], bins=21, align='left', color='g', edgecolor='red', linewidth=.8)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Histogram of iris flower dataset", fontdict = font1)
plt.xlabel("Species", fontdict = font2)
plt.ylabel("Number", fontdict = font2)
```

```python
plt.figure(figsize=(10,10))
sns.countplot(data1["Classification"])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Countplot of iris flower dataset", fontdict = font1)
plt.xlabel("Species", fontdict = font2)
plt.ylabel("Number", fontdict = font2)
```
